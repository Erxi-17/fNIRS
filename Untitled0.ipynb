```python
# STEP 1: INSTALL REQUIRED LIBRARIES (If not already installed)
!pip install pandas seaborn matplotlib lifelines

# STEP 2: IMPORT LIBRARIES
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from lifelines import CoxPHFitter, KaplanMeierFitter

# ------------------------------
# 🧠 BLOCK 1: REAL WORLD INPUT (Cerebral Tissue → Raw Data)
# ------------------------------
# Simulate input from cerebral tissue by using stroke dataset
df = pd.read_csv('s.csv')  # Upload your file here

# Display sample data
print("Sample Input Data:")
display(df.head())

# ------------------------------
# 💡 BLOCK 2: IR LEDs + PHOTODIODE MODULE (Data Acquisition)
# ------------------------------
# Feature selection simulating signal detection
features = ['age', 'hypertension', 'heart_disease', 'avg_glucose_level']
data = df[features + ['stroke']].dropna().copy()

# ------------------------------
# ⚙️ BLOCK 3: SIGNAL CONDITIONING CIRCUIT (Amplifier + Filter)
# ------------------------------
# Normalize signals (features) = Filtering Noise
data[features] = (data[features] - data[features].mean()) / data[features].std()

# Add survival_time (simulated) and stroke_event
np.random.seed(42)
data['survival_time'] = np.where(data['stroke'] == 1,
                                 np.random.randint(20, 300, len(data)),
                                 np.random.randint(300, 500, len(data)))
data['stroke_event'] = data['stroke']

# ------------------------------
# 📡 BLOCK 4: ESP32 MICROCONTROLLER (ADC + Feature Extraction)
# ------------------------------
# Perform survival analysis using CoxPH Model
cph = CoxPHFitter()
cph.fit(data, duration_col='survival_time', event_col='stroke_event')
print("📊 Feature Processing Summary:")
cph.print_summary()

# ------------------------------
# 🌐 BLOCK 5: OUTPUT INTERFACE (Cloud/Display)
# ------------------------------
# Visualize survival function using Kaplan-Meier
kmf = KaplanMeierFitter()
kmf.fit(durations=data['survival_time'], event_observed=data['stroke_event'])

plt.figure(figsize=(8, 5))
kmf.plot_survival_function()
plt.title("Stroke Patient Survival Curve")
plt.xlabel("Time (days)")
plt.ylabel("Survival Probability")
plt.grid(True)
plt.show()

# Predict survival risk scores
data['predicted_risk'] = cph.predict_partial_hazard(data)

# Display risk score histogram
plt.figure(figsize=(8, 4))
sns.histplot(data['predicted_risk'], kde=True, color='crimson')
plt.title("🔍 Predicted Stroke Risk Score Distribution")
plt.xlabel("Risk Score")
plt.grid(True)
plt.show()
```

```python
# ---------------------------------------------
# 🧠 COMSYS 2025 - Hardware-Mimicked Stroke Survival Analysis
# ---------------------------------------------

!pip install lifelines pandas seaborn matplotlib --quiet

# 📦 Import Modules
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from lifelines import CoxPHFitter, KaplanMeierFitter

# ---------------------------------------------
# 🧠 Step 1: Real World Input (Cerebral Tissue)
# Mapped as: Stroke Dataset Simulation
# ---------------------------------------------
df = pd.read_csv('s.csv')
print("Real-world input preview (simulated cerebral signals):")
print(df.head())

# ---------------------------------------------
# 💡 Step 2: IR LEDs + Photodiode Pair Module
# Mapped as: Biosignal capture - filtered columns
# ---------------------------------------------
selected_cols = ['age', 'hypertension', 'heart_disease', 'avg_glucose_level', 'stroke']
df = df[selected_cols].apply(pd.to_numeric, errors='coerce')
df.dropna(inplace=True)

# ---------------------------------------------
# ⚙️ Step 3: Signal Conditioning Circuit
# Mapped as: Signal normalization
# ---------------------------------------------
features = ['age', 'hypertension', 'heart_disease', 'avg_glucose_level']
df[features] = (df[features] - df[features].mean()) / df[features].std()

# ---------------------------------------------
# 📶 Step 4: ESP32 Microcontroller
# Mapped as: Feature Extraction + ADC + Risk Scoring
# ---------------------------------------------
np.random.seed(42)
df['survival_time'] = np.where(df['stroke'] == 1,
                               np.random.randint(50, 250, size=len(df)),
                               np.random.randint(250, 400, size=len(df)))
df['stroke_event'] = df['stroke'].astype(int)

cph = CoxPHFitter()
cph.fit(df[features + ['survival_time', 'stroke_event']],
        duration_col='survival_time', event_col='stroke_event')
df['predicted_risk'] = cph.predict_partial_hazard(df)

# ---------------------------------------------
# 🌐 Step 5: Output Interface (Cloud/Display)
# Mapped as: Risk Grouping and Data Export
# ---------------------------------------------
df['risk_group'] = pd.qcut(df['predicted_risk'], q=3, labels=['Low', 'Medium', 'High'])

# Export CSV and Summary
df.to_csv('stroke_risk_processed_comsys2025.csv', index=False)
with open("cox_model_summary_comsys2025.txt", "w") as f:
    cph.print_summary(print_func=lambda x: f.write(x + "\n"))

# ---------------------------------------------
# 📊 Step 6: Survival Model Insights
# Mapped as: Visualization (Kaplan-Meier + Risk Group)
# ---------------------------------------------

# 🧪 Correlation Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df[features].corr(), annot=True, cmap='coolwarm')
plt.title(" Feature Correlation (Simulated Cerebral Input)")
plt.tight_layout()
plt.show()

# 📈 Risk Group Distribution
plt.figure(figsize=(8, 4))
sns.countplot(x='risk_group', data=df, palette='viridis')
plt.title(" Risk Level Distribution (ESP32 Output)")
plt.grid(True)
plt.tight_layout()
plt.show()

# 🎻 Violin Plot of Risk Scores
plt.figure(figsize=(8, 5))
sns.violinplot(x='stroke_event', y='predicted_risk', data=df, palette='coolwarm')
plt.title(" Predicted Risk (Microcontroller Output vs Stroke Event)")
plt.xticks([0, 1], ['No Stroke', 'Stroke'])
plt.grid(True)
plt.tight_layout()
plt.show()

# ⏳ Kaplan-Meier Curves (Survival Insights)
plt.figure(figsize=(10, 6))
for group in df['risk_group'].unique():
    kmf = KaplanMeierFitter()
    group_data = df[df['risk_group'] == group]
    if not group_data.empty:
        kmf.fit(group_data['survival_time'], group_data['stroke_event'], label=f"Risk: {group}")
        kmf.plot_survival_function()

plt.title(" Survival Curves (Model Insights)")
plt.xlabel("Time (days)")
plt.ylabel("Survival Probability")
plt.grid(True)
plt.legend(title="Risk Group")
plt.tight_layout()
plt.show()

# 📘 Model Coefficients Plot
plt.figure(figsize=(8, 6))
cph.plot()
plt.title(" Feature Importance via Cox Model")
plt.grid(True)
plt.tight_layout()
plt.show()

print("✅ COMSYS 2025 pipeline complete: From sensor to survival model.")
```
